```python
import pandas as pd
ser1=pd.Series({'Material':'Wooden',
                'Colour':'Maroon',
                'Price': 25000})
ser2=pd.Series({'Material':'Plywood',
                'Colour':'Yellow',
                'Price': 20000})
ser3=pd.Series({'Material':'Plastic',
                'Colour':'Red',
                'Price': 1500})
ser4=pd.Series({'Material':'Stainless Steel',
                'Colour':'Silver',
                'Price': 55000})
ser5=pd.Series({'Material':'Wooden',
                'Colour':'Light Blue',
                'Price': 2500})
ser6=pd.Series({'Material':'Aluminium',
                'Colour':'Golden',
                'Price': 65000})
df=pd.DataFrame([ser1,ser2,ser3,ser4,ser5,ser6],index=['Sofa','Dinning Table','Chair','Sofa','Chair','Dinning Table'])
print(df)
print("\nChair and sofa:")
print(df.loc[['Chair','Sofa']])
print("\nFurniture Details:")
print(df.loc[df['Price']>25000])
print("\nFurniture Details:")
print(df.loc[df['Price']<1000])
print("\nAlternative Rows:")
print(df.iloc[0::2])
```

```python
import pandas as pd
brics=pd.read_csv('brics.csv',index_col=0)
print(brics)
print("\n",brics.describe())
print("\n\n",brics["country"])
brics['on_earth']='True'
print("\n\n",brics)
brics["density"]=brics["population"]/brics["area"]*1000000
print("\n\n",brics)
print("\n\n",brics.iloc[1:3])
print("\n\n",brics.iloc[1:3,0:2])
print("\n",brics["capital"].loc["CH"])
print("\n",brics.loc["CH"]["capital"])
print("\n",brics.loc["CH","capital"])
```

```python
import pandas as pd
#census=pd.read_csv('census.csv')
#print(census)
census_df=pd.read_csv('census.csv')
#print(census_df.groupby('STATE')['COUNTY'].count())
x=census_df.groupby('STNAME').count()['SUMLEV']
ans=x.idxmax()
print(ans)
```

```python
import pandas as pd
census_df=pd.read_csv('census.csv')
#x=census_df.groupby('COUNTY')['CENSUS2010POP'].nlargest(3)
x_df=census_df.sort_values(by=['STNAME','CENSUS2010POP'],ascending=False).groupby('STNAME').head(3)
ans=x_df.groupby('STNAME').sum().sort_values(by='CENSUS2010POP',).head(3).index.tolist()
print(ans)
```

```python
df=pd.read_csv('olympics.csv',index_col=0,skiprows=1)
df
```

```python
for col in df.columns:
    if col[:2]=='01':
        df.rename(columns={col:'Gold'+ col[4:]},inplace=True)
    if col[:2]=='02':
        df.rename(columns={col:'Silver'+ col[4:]},inplace=True)
    if col[:2]=='03':
        df.rename(columns={col:'Bronze'+ col[4:]},inplace=True)
    if col[:1]=='?':
        df.rename(columns={col:'#'+ col[1:]},inplace=True)
df.head()
```
