```python
import pandas as pd
census = pd.read_csv("census.csv")
census
```

```python
x= census.groupby("STNAME").count()['SUMLEV']
ans = x.idxmax()
ans
```

```python
top_counties_df = census.sort_values(by=['STNAME','CENSUS2010POP'], ascending=False).groupby("STNAME").head(3)
ans = top_counties_df.groupby('STNAME').sum().sort_values(by='CENSUS2010POP').head(3).index.tolist()
print(ans)
```

```python
import pandas as pd
df = pd.read_csv("olympics.csv")
df
```

```python
df = pd.read_csv("olympics.csv", index_col=0, skiprows=1)
df
```

```python
for col in df.columns:
  if col[:2]=='01':
    df.rename(columns={col:'Gold' + col[4:]}, inplace=True)
  if col[:2]=='02':
    df.rename(columns={col:'Silver' + col[4:]}, inplace=True)
  if col[:2]=='03':
    df.rename(columns={col:'Bronze' + col[4:]}, inplace=True)
  if col[:1]=='?':
    df.rename(columns={col:'#' + col[1:]}, inplace=True)
df.head()
```

```python
only_gold = df.where(df['Gold'] > 0)
only_gold.head()
```

```python
only_gold['Gold'].count()
```

```python
df['Gold'].count()
```

```python
only_gold = only_gold.dropna()
only_gold.head()
```
